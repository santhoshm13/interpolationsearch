```python
import random
import time

# ──────────────────────────────────────────────
# Interpolation Search
# ──────────────────────────────────────────────

def interpolation_search(arr, key):
    low, high = 0, len(arr) - 1
    probes = 0

    while low <= high and arr[low] <= key <= arr[high]:
        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            break

        pos = low + ((key - arr[low]) * (high - low) // (arr[high] - arr[low]))

        if arr[pos] == key:
            probes += 1
            return pos, probes
        elif arr[pos] < key:
            probes += 1
            low = pos + 1
        else:
            probes += 1
            high = pos - 1

    return -1, probes


# ──────────────────────────────────────────────
# Dataset Generators
# ──────────────────────────────────────────────

def generate_uniform(n):
    """Evenly spaced integers — ideal for interpolation search."""
    return sorted(random.sample(range(1, n * 10 + 1), n))


def generate_non_uniform(n):
    """Exponentially skewed values — worst case for interpolation search."""
    return sorted([int(random.expovariate(1 / 100)) for _ in range(n)])


# ──────────────────────────────────────────────
# User Input Helpers
# ──────────────────────────────────────────────

def get_int(prompt):
    while True:
        try:
            return int(input(prompt).strip())
        except ValueError:
            print("  ✗ Please enter a valid integer.")


def get_choice(prompt, valid):
    while True:
        ch = input(prompt).strip()
        if ch in valid:
            return ch
        print(f"  ✗ Enter one of: {', '.join(valid)}")


def get_dataset(label):
    print(f"\n--- {label} Dataset ---")
    mode = get_choice("  Enter elements  [1] Manually  [2] Randomly : ", ["1", "2"])

    if mode == "1":
        raw = input("  Enter sorted elements separated by spaces: ").strip().split()
        try:
            arr = [int(x) for x in raw]
        except ValueError:
            print("  ✗ Non-integer detected. Using empty list.")
            arr = []
        arr.sort()
        print(f"  Sorted dataset : {arr}")
    else:
        n = get_int("  How many elements to generate? : ")
        dist = get_choice("  Distribution  [1] Uniform  [2] Non-Uniform : ", ["1", "2"])
        arr = generate_uniform(n) if dist == "1" else generate_non_uniform(n)
        print(f"  Generated ({len(arr)} elements): {arr[:20]}{'...' if len(arr) > 20 else ''}")

    return arr


# ──────────────────────────────────────────────
# Search & Report
# ──────────────────────────────────────────────

def run_search(arr, key, label):
    start = time.perf_counter()
    index, probes = interpolation_search(arr, key)
    elapsed = (time.perf_counter() - start) * 1e6  # µs

    print(f"\n  [{label}] Search Results")
    print(f"  {'─'*35}")
    if index != -1:
        print(f"  Status      : ✔ FOUND")
        print(f"  Index       : {index}")
    else:
        print(f"  Status      : ✘ NOT FOUND")
    print(f"  Probes      : {probes}")
    print(f"  Time        : {elapsed:.4f} µs")

    return probes, elapsed


# ──────────────────────────────────────────────
# Comparison Summary
# ──────────────────────────────────────────────

def compare(u_probes, u_time, nu_probes, nu_time):
    print("\n" + "=" * 50)
    print("  COMPARISON SUMMARY")
    print("=" * 50)
    print(f"  {'Metric':<20} {'Uniform':>12} {'Non-Uniform':>12}")
    print(f"  {'─'*44}")
    print(f"  {'Probes':<20} {u_probes:>12} {nu_probes:>12}")
    print(f"  {'Time (µs)':<20} {u_time:>12.4f} {nu_time:>12.4f}")
    print("=" * 50)

    if u_probes < nu_probes:
        print("  ✔ Uniform data needed FEWER probes — as expected.")
    elif u_probes > nu_probes:
        print("  ✔ Non-uniform data happened to need fewer probes this run.")
    else:
        print("  Both datasets needed the same number of probes.")

    print("""
  Key Insight:
  Interpolation Search estimates the probe position using
  the formula:
    pos = low + (key - arr[low]) * (high - low)
              / (arr[high] - arr[low])
  This works best when values are UNIFORMLY spread, giving
  O(log log n) average comparisons.  On SKEWED/exponential
  data the estimate is poor, probes increase, and worst-case
  degrades to O(n).
""")


# ──────────────────────────────────────────────
# Main
# ──────────────────────────────────────────────

def main():
    print("=" * 50)
    print("  INTERPOLATION SEARCH — UNIFORM vs NON-UNIFORM")
    print("=" * 50)

    # ── Uniform dataset ──
    u_arr = get_dataset("UNIFORM")
    if not u_arr:
        print("  Empty dataset — skipping uniform search.")
        u_probes, u_time = 0, 0.0
    else:
        u_key = get_int("\n  Enter search key for UNIFORM dataset : ")
        u_probes, u_time = run_search(u_arr, u_key, "UNIFORM")

    # ── Non-uniform dataset ──
    nu_arr = get_dataset("NON-UNIFORM")
    if not nu_arr:
        print("  Empty dataset — skipping non-uniform search.")
        nu_probes, nu_time = 0, 0.0
    else:
        nu_key = get_int("\n  Enter search key for NON-UNIFORM dataset : ")
        nu_probes, nu_time = run_search(nu_arr, nu_key, "NON-UNIFORM")

    # ── Compare ──
    compare(u_probes, u_time, nu_probes, nu_time)


if __name__ == "__main__":
    main()
```
